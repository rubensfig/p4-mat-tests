```python
import os
import sys
import pandas as pd
import json
import re
```

```python
results_file = 'results.json'
```

```python
sram_dp = pd.DataFrame()
tcam_dp = pd.DataFrame()
for root, subdirs, files in os.walk('.'):
    if results_file in files:
        js = pd.read_json(root +'/' + results_file)
        number = re.search(r"./(.*)-results-\d*-(\d*).*", root)
        #print(number.groups())
        t = js['latency'].dropna()[1]['7']['latency'] | {'n_clients': number.groups()[1]} | {'type':  number.groups()[0]}
        tcam_dp = tcam_dp.append(t, ignore_index=True)
```

```python
q_low = tcam_dp["total_max"].quantile(0.01)
q_hi  = tcam_dp["total_max"].quantile(0.99)

dp_filtered = tcam_dp[( tcam_dp["total_max"] < q_hi) & ( tcam_dp["total_max"] > q_low)]
```
